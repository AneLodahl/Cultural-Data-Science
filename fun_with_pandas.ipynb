## Music sales challenge

Write a script that:

1. Combines the tables of best-selling physical singles and best-selling digital singles on the Wikipedia page "List_of_best-selling_singles"
2. Adds a column which marks whether each row is from the list of physical singles or digital singles
3. Outputs the artist and single name for the year you were born. If there is no entry for that year, take the closest year after you were born.
4. Outputs the artist and single name for the year you were 15 years old.

```python
# Starter code...

%pip install lxml
import pandas as pd

rawdata = pd.read_html("https://en.wikipedia.org/wiki/List_of_best-selling_singles")
```

```python
#table for physical
df_physical = rawdata[0]
df_physical
```

```python
#add column for physical

df_physical['Phy/Dig'] = pd.Series(['Physical' for x in range(len(df_physical.index))])
df_physical
```

```python
df_digital = rawdata[3]
df_digital
```

```python
#add column for digital
df_digital['Phy/Dig'] = pd.Series(['Digital' for x in range(len(df_digital.index))])
df_digital

#why suddenly the na??
```

```python
df_combined = pd.concat([df_physical, df_digital], axis=0)
df_combined
```

```python
#finding release at 2000
df_combined['Released'] == 2000

#closest to is "Something About the Way You Look Tonight" by Elton John	from 1997
```

```python
#seeing if I can also do it if it was a longer list

#find closest to 2000 and save the index
df_combined['diff_from_2000'] = abs(df_combined['Released'] - 2000)
```

```python
#find the song with the smallest difference from 2000
min_diff_index = df_combined['diff_from_2000'].idxmin()
```

```python
#get the column
df_combined.loc[min_diff_index, ['artist', 'song_title']]
```

```python
#same for year 2015

df_combined['Released'] == 2015

#gives true at 5, 8, and 10
#which are
#Wiz Khalifa featuring Charlie Puth	"See You Again"
#Mark Ronson featuring Bruno Mars	"Uptown Funk"
#Ed Sheeran	"Thinking Out Loud"
```

## Space challenge

1. Make a single dataframe that combines the space missions from the 1950's to the 2020's
2. Write a script that returns the year with the most launches
3. Write a script that returns the most common month for launches
4. Write a script that ranks the months from most launches to fewest launches


```python
# Starter code...

%pip install lxml
import pandas as pd

rawdata_space = pd.read_html("https://en.wikipedia.org/wiki/Timeline_of_Solar_System_exploration")
```

```python
df_space_combined = pd.concat(rawdata_space[0:8])
df_space_combined
```

```python
#create new column with launch year
df_space_combined['Launch_year'] = df_space_combined['Launch date'].str.extract(r'(\d{4})')
df_space_combined
```

```python
#most launches, year
df_space_combined['Launch_year'].value_counts(normalize=False)



```

```python
most_launches_year = df_space_combined['Launch_year'].mode()
most_launches_year
```

```python
#most launches, month

df_space_combined['Launch_month'] = df_space_combined['Launch date'].str.extract(r'([A-Za-z]+)')
df_space_combined
```

```python
df_space_combined['Launch_month'].value_counts(normalize=False)
```
